```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy
import re
```

```python
#Open the text file
henryHTML = open('./10KQ/HENRY_SCHEIN_INC.txt', 'r+')
##henryHTML = open('./10KQ/MEDALLION_FINANCIAL_CORP.txt', 'r+')
##henryHTML = open('./10KQ/DISH_Network_CORP.txt', 'r+')


soup = henryHTML.read()
bs = BeautifulSoup(soup,"html.parser")
#print(type(bs))

#Remove ASCII-Encoded segments – All document segment <TYPE> tags of GRAPHIC, ZIP, EXCEL, JSON, and PDF
segments_to_be_removed = ['GRAPHIC', 'ZIP', 'EXCEL', 'JSON', 'PDF']

for tag in bs.find_all('type'):
    newtag = str(tag.find(text=True))
    if (newtag.rstrip() in segments_to_be_removed):
        tag.parent.decompose()
        

#Extract the SEC header text     
headerText = ''

for header in bs.findAll('sec-header'):
    headerText += header.text
    

    
# Extract the list of filing types
typeList = []
for val in bs.find_all('type'):
    typeList.append(val.find(text=True).rstrip())
    
 
```

```python
def removeTags(tagToRemove):
    for tag in bs.find_all(tagToRemove):
        tag.decompose()
        
#Removing all <TR>, <TD>, <FONT>, and <XBRL> tags
tags_to_be_removed = ['tr', 'td', 'font', 'xbrl', 'table', 'presentationlink', 'sec-header', 
                      'img', 'hr']        

for tags in tags_to_be_removed:
    removeTags(tags)



## Get the list of tags from which contents to be extracted
invalid_tags = []
for tag in bs.find_all():
    invalid_tags.append(tag.name)
    
invalid_tags = set(invalid_tags) 
print(invalid_tags)

```

```python
## Remove all the table headers
for a in bs.find_all('a', href=True):
    if a['href'] == '#TABLEOFCONTENTS':
        a.decompose()
        

## Remove the footers
for footer in bs.find_all('div', bclfooter = True):
    footer.decompose()
```

```python
## Extract text from all the corresponding tags
for tag in invalid_tags: 
    for match in bs.findAll(tag):
        match.replaceWithChildren()
```

```python
## Remove all the white spaces
stringExtracted = str(bs)
stringExtracted = re.sub('\s+', ' ',stringExtracted)
print(stringExtracted)
```

```python
## write the output to a file "output_henry"
with open("output_henry.txt",'w', encoding='utf-8') as file:
    file.write(stringExtracted) 
```
